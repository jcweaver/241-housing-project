```python
import pandas as pd
import numpy as np
```

```python
temp = pd.read_csv("data/tracts_data_hisp.csv")
```

```python
#Clean up some of the rows that contian missing values
# -666666666 is missing and should remove this row
df = temp.copy()
for col in temp.columns:
    df = df.loc[df[col]!=-666666666]
```

```python
df.columns
```

```python
#Let's look for county pairings
states = np.unique(df.state)
states
```

```python
#Set a list of eval_columns to use for evaluation
eval_columns = ['population_density', 'housing_density', 'median_age',
       'median_male_age', 'median_income',  'pct_hs_grad', 'NatWalkInd']
```

```python
#Closeness calculation
f = .0175 #Resulted in 0 when .01 #resulted in 200,558 when .1
def calculation(column_name):
    #return (np.max(df[column_name])-np.min(df[column_name]))*f
    return np.std(df[column_name])*.125
```

```python
#Use a dictionary to store the std dev & mean for each column
eval_dict = {}

for col in eval_columns:
    eval_dict[col] = calculation(col)
    
eval_dict
```

```python
#Use a dictionary to store the std dev & mean for each column
eval_dict = {}

for col in eval_columns:
    eval_dict[col] = calculation(col)
    
eval_dict
```

```python
pairings = [] #List of tuples to store GEOIDS of possible pairings

for state in states[2:]:
    temp = df.loc[df['state']==state]
    counties = np.unique(temp.county)
    
    for county in counties:
        temp2 = temp.loc[temp['county']==county]
        
        #Find the unique tracts in this county & state
        ids = np.unique(temp2.GEOID)
    
        #Some optimization since I don't need to double check pairs that have already been tried
        for i in range(len(ids)-1):
            a = ids[i]
            for j in range(i+1, len(ids)-1):
                b = ids[j]
                
                #First need to confirm geoids are racially different
                if ((temp2.loc[temp2['GEOID']==a, "pct_white"].item() < .5)
                and (temp2.loc[temp2['GEOID']==b, "pct_white"].item() > .5)) or ((temp2.loc[temp2['GEOID']==a, "pct_white"].item() > .5)
                and (temp2.loc[temp2['GEOID']==b, "pct_white"].item() < .5)):
     

                    #a and b are two geoids which can be used to find a tract
                    #Here is a potential comparison we can use to find pairs
                    for col in eval_columns:
                        #print(temp2.loc[temp2['GEOID']==a, col].item())
                        if abs(temp2.loc[temp2['GEOID']==a, col].item()-temp2.loc[temp2['GEOID']==b, col].item()) <= eval_dict[col]:
                            continue
                        else:
                            break
                    else:
                        pairings.append((a,b))
```

```python
len(pairings)
```

```python
print(pairings)
```

```python
newdf = pd.DataFrame()
paircount = 0
for pair in pairings:
    for geoid in pair:
        newdf = newdf.append(df.loc[df['GEOID']==geoid])
        newdf.loc[newdf['GEOID']==geoid, "Pair_ID"] = paircount
    paircount+=1
```

```python
newdf.to_csv("pairings_26.csv")
```
